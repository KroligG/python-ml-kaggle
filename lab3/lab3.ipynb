```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from patsy import dmatrices
import re
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.decomposition import PCA
import sklearn.preprocessing as preprocessing
```

```python
df = pd.read_csv("data/train.csv")
df['target_n'] = pd.factorize(df.target)[0]

df_test = pd.read_csv("data/test.csv")
sampleSubmission = pd.read_csv("data/sampleSubmission.csv")

df.head()
```

```python
classifier = GradientBoostingClassifier(n_estimators=400)
classifier.fit(df[df.columns[1:-2]], df.target_n)
preds = classifier.predict_proba(df_test[df_test.columns[1:]])
result = pd.DataFrame(preds, index=df_test.id.values, columns=sampleSubmission.columns[1:])
result.head()
```

```python
result.to_csv('result_gb.csv', index_label='id')
```
